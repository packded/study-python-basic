```python
set_df.PBV.plot()
#data ตรงหลุมก็ดูแปลกๆ เราอาจต้องไปดูว่าเกิดอะไรขึ้น ณ ช่วงเวลานั้นหรือ ข้อมูลมันผิด
```

```python
AOTtoSetpb = df.PBV/set_df.PBV 
AOTtoSetpb #ข้อมูลเรียงจากน้อยมามากอยู่แล้ว
```

```python
AOTtoSetpb.plot()
#ไอแท่งที่โด่ขึ้นมานั้นอาจเป็น outlier เราอาจไปตัดออก
```

```python
#เราจะตัด outlier ด้วยใช้ข้อมูลแค่ถึง2018อะไรงี้

AOTtoSetpb.loc[:'2018'].plot() 
```

```python
data2 = AOTtoSetpb.loc[:'2018']
df2 = pd.DataFrame(AOTtoSetpb.loc[:'2018'])
df2
```

```python
#ถ้าข้อมูลยังเรียงจากน้อยไปมาก เรียงจากมากไปน้อยอยู่วิธีแก้คือ 
bband = pd.DataFrame(AOTtoSetpb.loc['2018':]).sort_index(ascending=True) #ตัดช่วงที่ต้องการมาแล้วมาเรียง
bband #อันนี้ทำเป้นตัวอย่างเฉยๆ 
```

```python
#กลับมาช่วงที่เราใช้จริง
mean = df2.PBV.ewm(span=60,adjust=False,min_periods=60).mean()
std  = df2.PBV.ewm(span=60,adjust=False,min_periods=60).std()
```

```python
import matplotlib.pyplot as plt
df2.plot(color='k')
mean.plot(color='r')
plt.fill_between(std.index,mean -2 * std,mean + 2*std,color='b',alpha=0.1)
```

เราจะสังเกตุเห็นควาสัมพันว่าถ้าชนเส่น mean จะเด้งลง เราอาจไปประบ
ค่าพารามิเตอต่างๆให้เกิดความแม่นยำขึ้น

```python
mean = df2.PBV.ewm(span=100,adjust=False,min_periods=100).mean()
std  = df2.PBV.ewm(span=100,adjust=False,min_periods=100).std()

df2.plot(color='k')
mean.plot(color='r')
plt.fill_between(std.index,mean -2 * std,mean + 2*std,color='b',alpha=0.1)
```
