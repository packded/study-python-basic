```python
import yfinance as yf
import pandas as pd 
import mplfinance as mpf
```

```python
data = yf.download('spy',start='2019-01-01',end='2023-01-01')
data
```

```python
df = data.copy()
```

```python
#ทำข้อมูลจาก daily เป็น weekly 
df.head(20)
```

```python
df.resample('W-Mon').last() #last() ดึงสุดท้ายเท่าที่มีมา
```

จะเห็นว่าการ resample แบบนี้ ข้อมูลมันมาเหมือนตัดมาของวันที่เราต้องการแต่มันไม่ใช่แท่ง 1Week แบบในกราฟที่เราต้องการจริงๆ

ดูจากข้างบนจะเห็นว่า open high low close ก็เป็นของวันที่ 4 เลยไม่มช่การรวมของข้อมุลทั้งสัปดาห์

```python
#ตัวอย่างที่ถูกต้องสำหรับการ resample  ต้องมี aggreegate ด้วย
df_weekly = df.resample('W-FRI').agg({ #ปกติจะ resammple weekly ที่วันศุกร์
                                'Open':'first', # first value of the week
                                'High':'max', # highest value of the week
                                'Low':'min', # lowest value of the week
                                'Close':'last', #last value of the week
                                'Volume':'sum'})#sum volume of the week
```

```python
df_weekly
```

```python
#ทำเป็น monthly โดยเอาแค่ business day (BM)
df_monthly = df.resample('BM').agg({
                                    'Open':'first',
                                    'High':'max',
                                    'Low':'min',
                                    'Close':'last',
                                    'Volume':'sum'
})
df_monthly
```

สิ่งที่ต้องระวังคือบางทีเราจะเอาตัวเลขทางเศรษฐกิจมาเทียบกับราคานะ บาทีมันเอามาเทียบกกันไม่ได้เพราะเลขทางพวกนี้ประกาศกลางเดดือนแต่ ราคาออกมาท้ายเดือน อาจจะมองไม่เห็นอะไร

```python
df_everyday = df.resample('D').last()
```

```python
df_everyday #จะเห็นว่ามี nan ในบางอันนั้นคือวันหยุด
```

```python
#การแก้ข้อมูลที่เป็น na ทำ 2 ขั้นตอนคือแบบนั้นคือ เพิ่มเข้ามาให้เต็มแล้วตัดออก

#ขั้นแรก  เพิ่มเข้ามา
#สิ่งที่จะทำก็คือเราจะเติมช่องว่างด้วย forward fill(ffill) - ก็คือเติมแบบวันนี้ไม่มีก็เอาวันก่อนหน้ามาเติม
df_everyday.fillna(method='ffill',inplace=True)
df_everyday
```

```python
#ขั้นที่สอง  ตัดออก
#สิ่งที่จะทำก็คือเราจะจัดวันที่ออกให้เหลือแต่ business day
df_businessday = df_everyday.resample('B').last()
df_businessday
```

สมมุติจะทำราคาเพื่อ predict week ถัดไปโดยใช้ OHLCV  (รู้ข้อมูลอาทิตย์นี้ แล้วอาทิตย์หน้าผลลัพธ์คืออะไรน้า?)

```python
df_weekly['Next week closing'] = df_weekly['Close'].shift(-1)
df_weekly    
```
