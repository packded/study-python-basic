```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y)
plt.ylabel('some numbers')
plt.xlabel('some value')
plt.show()
```

```python
x = np.linspace(0,2,100)
x
```

```python
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.plot(x,2**x,label='expo')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend();
```

```python
x= np.array([130,147,138,185,174,166,153,162,168])
y= np.array([96,73,88,92,81,70,69,85,84])

plt.xlim(120,200)
plt.ylim(60,100)
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('Height')
plt.ylabel('weight')
plt.show()
```

```python
#barplot
x = ['bar-1','bar-2','bar-3','bar-4']
y= [70,65,30,58]

plt.bar(x,y,color='red')
plt.title('Bar plot')
plt.xlabel('Type')
plt.ylabel('Range')
plt.show()
```

```python
#barplot stack
manu = ['fire rice','kapao','noodle','egg boiled']
boy= [70,65,30,58]
girl = [10,10,10,20]

plt.bar(manu,boy,color='k')
plt.bar(manu,girl,color='cyan',bottom=boy)
plt.title('Bar plot')
plt.xlabel('manu')
plt.ylabel('amount')
plt.
plt.show()
```

```python
#histogram

x = np.random.randn(10000)

plt.title('Histogram')
plt.xlabel('Random data')
plt.ylabel('Freq')
plt.hist(x,50)
plt.grid()
plt.show()
```

```python
#subplot
#marker    https://matplotlib.org/stable/api/markers_api.html
a=[1,2,3,4,5]
b=[0,0.6,0.2,15,10,8,6,21]
c=[4,2,6,8,3,20,13,15]

fig = plt.figure(figsize=(20,14))

sub1 = plt.subplot(2,2,1)
sub2 = plt.subplot(2,2,2)
sub3 = plt.subplot(2,2,3)
sub4 = plt.subplot(2,2,4)

sub1.plot(a,'sb')
sub1.set_xticks(list(range(0,10,1)))
sub1.set_title('1st')

sub2.plot(b,'or')
sub2.set_xticks(list(range(0,10,2)))
sub2.set_title('2nd')

sub3.plot(list(range(0,22,3)),'vg')
sub3.set_xticks(list(range(0,10,1)))
sub3.set_title('3rd')

sub4.plot(c,'Dm')
sub4.set_xticks(list(range(0,24,2)))
sub4.set_title( '4th')

plt.show()
```
