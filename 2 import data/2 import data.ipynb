```python
from pandas_datareader import data as pdr  # import ตรงนี้สำคัญนะจะขึ้นไม่ขึ้น
import pandas as pd 
import yfinance as yf
yf.pdr_override()  #must have this line for using yf!!
```

```python
start = '2019-01-01'
end   = '2023-12-31'
```

```python
tsla = pdr.get_data_yahoo('TSLA', start=start, end=end)
```

```python
tsla
```

```python
#ทำข้อมูล daily > weekly
tsla_weekly = tsla.asfreq('W-Mon')
tsla_weekly
```

```python
#ใช้คำสั่ง resample มาทำก็ได้อีกทาง
tsla_monthly = tsla.resample('M').mean()
tsla_monthly
```
